```python
# @title Шифровка по Азбуке Морзе
# Задание №1*
morze = {'a': '.-', 'b': '-…', 'c': '-.-.', 'd': '-..',
 'e': '.', 'f': '..-.', 'g': '--.', 'h': '….',
 'i': '..', 'j': '.---', 'k': '-.-', 'l': '.-..',
 'm': '--', 'n': '-.', 'o': '---', 'p': '.--.',
 'q': '--.-', 'r': '.-.', 's': '…', 't': '-',
 'u': '..-', 'v': '…-', 'w': '.--', 'x': '-..-',
 'y': '-.--', 'z': '--..'}

decode_message = input()
encode_message = ""

for letter in decode_message:
  if letter == ' ':
    encode_message += "\n"
  else:
    encode_message += morze[letter.lower()] + " "

print(encode_message)
```

```python
# @title Система регистрации пользователей
# Задание №2*
# fpqhfouqdldravpjttarh
# jmvlplnrmba
user_db = []
names = []

for i in range(0, int(input())):
  names.append(input())

print("")
for name in names:
  if name not in user_db:
    user_db.append(name)
    print("OK")
  else:
    i = 1
    while f"{name}{i}" in user_db:
      i += 1
    user_db.append(f"{name}{i}")
    print(f"{name}{i}")
```

```python
# @title Система обработки запросов пользователей файловой системы
# Задание №3*
# Представим схему доступа в виде таблицы
# r w x number
# 0 0 0 0
# 0 0 1 1
# 0 1 0 2
# 0 1 1 3
# 1 0 0 4
# 1 0 1 5
# 1 1 0 6
# 1 1 1 7
# Соотвественно после получения каждой буквы мы увеличиваем уровень доступа
# на x(+1)/w(+2)/r(+4) таким образом можно однозначно понять что разрешено,
# а что нет:
# execute - при 1 или 3 или 5 или 7
# write - при 2 или 3 или 6 или 7
# read - при 4 или 5 или 6 или 7

def send_command(command, permission):
  if command == "execute":
    if permission in (1, 3, 5, 7):
      print("OK")
    else:
      print("Access denied")
  if command == "write":
    if permission in (2, 3, 6, 7):
      print("OK")
    else:
      print("Access denied")
  if command == "read":
    if permission in (4, 5, 6, 7):
      print("OK")
    else:
      print("Access denied")

files_db = {}
query_list = []
permission_logic = {"r": 4, "w": 2, "x": 1}

for i in range (0, int(input())):
  file_info = input().split(" ")
  if len(file_info) == 4:
    files_db[file_info[0]] = 7
  elif len(file_info) == 3:
    files_db[file_info[0]] = permission_logic[file_info[1]] + \
                             permission_logic[file_info[2]]
  else: # == 2
    files_db[file_info[0]] = permission_logic[file_info[1]]
for i in range (0, int(input())):
  query_list.append(input())

print("")
for query in query_list:
  command = query.split(" ")[0]
  permission = files_db[query.split(" ")[1]]
  send_command(command,permission)
```
