```python
# Задание №1*
morze = {'a': '.-', 'b': '-…', 'c': '-.-.', 'd': '-..',
 'e': '.', 'f': '..-.', 'g': '--.', 'h': '….',
 'i': '..', 'j': '.---', 'k': '-.-', 'l': '.-..',
 'm': '--', 'n': '-.', 'o': '---', 'p': '.--.',
 'q': '--.-', 'r': '.-.', 's': '…', 't': '-',
 'u': '..-', 'v': '…-', 'w': '.--', 'x': '-..-',
 'y': '-.--', 'z': '--..'}

decode_message = input()
encode_message = ""

for letter in decode_message:
  if letter == ' ':
    encode_message += "\n"
  else:
    encode_message += morze[letter.lower()] + " "

print(encode_message.strip())
```
