```python
# @title Блок импортов

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import random as rnd
import math
```

```python
# @title Задание 1 Загрузить данные из файла “insurance.csv”.

insurance = pd.read_csv('/content/insurance.csv')
insurance.info()
```

```python
# @title Задание 3 Построить гистограммы для числовых показателей.

insurance.hist(edgecolor='black', color='green', figsize=(15,6))
```

```python
# @title Задание 4 Найти меры центральной тенденции и меры разброса для индекса массы тела (bmi) и расходов (charges). Отобразить результаты в виде текста и на гистограммах (3 вертикальные линии). Добавить легенду на графики.

# Меры центральной тенденции: Мода, Медиана, Среднее (moda, median, mean)
# Меры разброса: Размах, Стандартное отклонение, Межквартальный размах

figure, axis = plt.subplots(1,2)
figure.set_figheight(10)
figure.set_figwidth(15)

# Для bmi
modaBMI = sts.mode(insurance.bmi)
medianBMI = np.median(insurance.bmi)
meanBMI = np.mean(insurance.bmi)
razBMI = insurance.bmi.max() - insurance.bmi.min()
stdBMI = insurance.bmi.std()
iqrBMI = sts.iqr(insurance.bmi, interpolation='midpoint')

print("Для bmi")
print(f"Мода: {modaBMI}\nМедиана: {medianBMI}\nСреднее: {meanBMI}")
print(f"Размах: {razBMI}\nСтандартное отклонение: {stdBMI}\nМежквартальный размах: {iqrBMI}")

axis[0].hist(insurance.bmi, color='green', edgecolor='black')
axis[0].axvline(x=modaBMI.mode, color='red', linestyle='dashed', linewidth=2, label='Мода')
axis[0].axvline(x=meanBMI, color='blue', linestyle='dashed', linewidth=2, label='Среднее')
axis[0].axvline(x=medianBMI, color='yellow', linestyle='dashed', linewidth=2, label='Медиана')
axis[0].legend()

# Для charges
modaCHARGES = sts.mode(insurance.charges)
medianCHARGES = np.median(insurance.charges)
meanCHARGES = np.mean(insurance.charges)
razCHARGES = insurance.charges.max() - insurance.charges.min()
stdCHARGES = insurance.charges.std()
iqrCHARGES = sts.iqr(insurance.charges, interpolation='midpoint')


print('*'*80)
print("Для bmi")
print(f"Мода: {modaCHARGES}\nМедиана: {medianCHARGES}\nСреднее: {meanCHARGES}")
print(f"Размах: {razCHARGES}\nСтандартное отклонение: {stdCHARGES}\nМежквартальный размах: {iqrCHARGES}")

axis[1].hist(insurance.charges, color='green', edgecolor='black')
axis[1].axvline(x=modaCHARGES.mode, color='red', linestyle='dashed', linewidth=2, label='Мода')
axis[1].axvline(x=meanCHARGES, color='blue', linestyle='dashed', linewidth=2, label='Среднее')
axis[1].axvline(x=medianCHARGES, color='yellow', linestyle='dashed', linewidth=2, label='Медиана')
axis[1].legend()

```

```python
# @title Задание 6 Используя признак charges или imb, проверить, выполняется ли центральная предельная теорема. Использовать различные длины выборок n. Количество выборок = 300. Вывести результат в виде гистограмм. Найти стандартное отклонение и среднее для полученных распределений.

# Проверка выполнения центральной предельной теоремы, выборок - 300, 1338 элементов
n = 300
size = [50, 100, 150]
maxEndSize = [1287, 1237, 1187]
means = []

figure, axis = plt.subplots(3,1)
figure.set_figheight(10)
figure.set_figwidth(6)

for j in range(0,3):
  for i in range(0,n):
    startPoint = rnd.randint(0,maxEndSize[j])
    group = insurance.bmi[startPoint:startPoint+size[j]]
    means.append(np.mean(group))

  axis[j].hist(means, 300, histtype='stepfilled')
  axis[j].set_title(f'bmi size={size[j]}')

  # Станадртное и среднее отклонение
  std = np.std(means)
  mean = np.mean(means)
  print(f'Для размера выборки {size[j]} стандартное отклонение - {std}, а среднее - {mean}')
```

```python
# @title Задание 7 Построить 95% и 99% доверительный интервал для среднего значения расходов и среднего значения индекса массы тела.

charges95 = sts.norm.interval(confidence=0.95, loc=np.mean(insurance.charges), scale=sts.sem(insurance.charges))
charges99 = sts.norm.interval(confidence=0.99, loc=np.mean(insurance.charges), scale=sts.sem(insurance.charges))
bmi95 = sts.norm.interval(confidence=0.95, loc=np.mean(insurance.bmi), scale=sts.sem(insurance.bmi))
bmi99 = sts.norm.interval(confidence=0.99, loc=np.mean(insurance.bmi), scale=sts.sem(insurance.bmi))

print(f'\n95% доверительный интервал для среднего значения расходов = {charges95}',
      f'\n99% доверительный интервал для среднего значения расходов = {charges99}',
      f'\n95% доверительный интервал для среднего значения индекса массы тела = {bmi95}',
      f'\n99% доверительный интервал для среднего значения индекса массы тела = {bmi99}')
```

```python
# @title Задание 8 Проверить распределения следующих признаков на нормальность: индекс массы тела, расходы. Сформулировать нулевую и альтернативную гипотезы. Для каждого признака использовать KS-тест и q-q plot.

# KS-тест
statBMI = sts.kstest(insurance.bmi, 'norm')
statCHARGES = sts.kstest(insurance.charges, 'norm')
print(f'KS-тест для индекса массы тела - {statBMI}',
      f'\nKS-тест для расходов - {statCHARGES}')

# Q-Q-plot
_, axis= plt.subplots(1,2,figsize=(8,4))
sts.probplot(insurance.bmi, plot=axis[0])
axis[0].set_title("Индекс массы тела")
sts.probplot(insurance.charges, plot=axis[1])
axis[1].set_title("Расходы")
```

```python
# @title Задание 13 Загрузить данные из файла “bmi.csv”. Взять оттуда две выборки. Одна выборка – это индекс массы тела людей c региона northwest, вторая  выборка – это индекс массы тела людей с региона southwest. Сравнить средние значения этих выборок, используя t-критерий Стьюдента. Предварительно проверить выборки на нормальность (критерий ШопироУилка) и на гомогенность дисперсии (критерий Бартлетта).

bmi = pd.read_csv("/content/bmi.csv")

bmiNorth = bmi[bmi.region == 'northwest']
bmiSouth = bmi[bmi.region == 'southwest']

print(f'Проверка на нормальность, критерий Шопиро-Уилка:\n  С севера - {sts.shapiro(bmiNorth.bmi)}\n  С юга - {sts.shapiro(bmiSouth.bmi)}, обе выборки нормальные')
print(f'Проверка на гомогенность дисперсии, критерий Бартлетта - {sts.bartlett(bmiNorth.bmi, bmiSouth.bmi)}, гомогенны')
print(f'Сравнение средних значений по t-критерию Стьюдента - {sts.ttest_ind(bmiNorth.bmi, bmiSouth.bmi)}, среднее отличаются')
```
